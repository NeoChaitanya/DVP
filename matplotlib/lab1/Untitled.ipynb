Line Plots


```python
import matplotlib.pyplot as plt
```

```python
line1 = [1,5,2,3,1,6,3,2]
line2 = [5,2,5,2,0,9,2,5]
```

```python
plt.plot(line1, label="line 1")
plt.plot(line2, label="line 2")

plt.plot()
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.title("Line graph")

plt.legend()
plt.show()
```

Bar plots

```python
x1 = [1,3,4,5,6,7,9]
y1 = [4,7,2,4,7,8,3]
x2 = [2,4,6,8,10]
y2 = [5,6,2,2,4]

plt.bar(x1, y1, label="bar 1", color='b')
plt.bar(x2, y2, label="bar 2", color='r')
plt.plot()

plt.xlabel("bar number")
plt.ylabel("bar height")
plt.title("bar chart")

plt.legend()
plt.show()
```

Histograms

```python
import numpy as np
```

```python
n = 5 + np.random.randn(1000)
m = [m for m in range (len(n))]

plt.bar(m,n)
plt.title("Raw data")
plt.show()
plt.hist(n, bins=20)
plt.title("histogram")
plt.show()

plt.hist(n, cumulative=True, bins=20)
plt.title("Cumulative histogram")
plt.show
```

# Scatter plot

```python
x1 = [2,4,5]
y1 = [5,5,6]
x2 = [1,2,6,3,6,8]
y2 = [2,6,8,9,4,3]
y3 = [3,6,3,1,9,5]

plt.scatter(x1, y1)
plt.scatter(x2, y2, marker='^', color='g')
plt.scatter(x2, y3, marker='v', color='b')

plt.title("Scatter plot")
plt.show()
```

# stack plts


```python
indexs = [1,2,3,4,5,6,7,8,9]
array1 = [23,33,66,43,29,53,11,55,75,]
array2 = [35,32,83,72,11,64,95,22,86]
array3 = [22,11,34,74,97,32,54,75,66]

plt.plot([], [], color='r', label='plot1')
plt.plot([], [], color='g', label='plot2')
plt.plot([], [], color='b', label='plot2')

plt.stackplot(indexs, array1, array2, array3, colors=['r','g','b'])

plt.title("Stack plot")

plt.legend()
plt.show()
```

# pie charts

```python
labels = 'Sl1', 'Sl2', 'Sl3', 'Sl4', 'Sl5'
sections = [22,21,11,9,37]

colors = ['c','y','g','b','r']

plt.pie(sections, labels=labels, colors=colors)

plt.axis('equal')
plt.title("Pie chart")
plt.show()
```
